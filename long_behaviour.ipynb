```python
import numpy as np
import matplotlib.pyplot as plt

h= 1e-5
fig = plt.figure(figsize=(10, 5))

E_AB = np.loadtxt('Adams-BashforthE.txt', unpack=True)
E_AM = np.loadtxt('Adams-MoultonE.txt', unpack=True)
E_V = np.loadtxt('VerletE.txt', unpack=True)
E_RK4 = np.loadtxt('RK4NE.txt', unpack=True)
E_FR = np.loadtxt('FRE.txt', unpack=True)
t_AB = np.linspace(0, len(E_AB)*h, len(E_AB))
t_AM = np.linspace(0, len(E_AM)*h, len(E_AM))
t_V = np.linspace(0, len(E_V)*h, len(E_V))
t_RK4 = np.linspace(0, len(E_RK4)*h, len(E_RK4))
t_FR = np.linspace(0, len(E_FR)*h, len(E_FR))

def rel_E(E):
    return np.abs((E-E[0])/E[0])

plt.plot(t_V, rel_E(E_V), linewidth=0.5, color='C0', label='Verlet', alpha=0.5)
plt.plot(t_FR, rel_E(E_FR), linewidth=0.5, color='orange', label='Forest-Ruth', alpha=0.7)
plt.plot(t_AB, rel_E(E_AB), color = 'green', label='Adams-Bashforth')
plt.plot(t_AM, rel_E(E_AM), color = 'purple', label='Adams-Moulton')
plt.plot(t_RK4, rel_E(E_RK4), color='red', label='Runge-Kutta 4')

plt.title('Long-term behaviour integrators (h=1e-5)')
plt.ylabel('|$\Delta$E/E|')
plt.xlabel('t (2.9722e9 s)')
plt.legend(loc='upper right')
plt.yscale('log')
#plt.ylim(10**(-17), 10**(-1))
plt.show()
```
