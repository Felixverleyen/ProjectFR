```python
import numpy as np
import matplotlib.pyplot as plt

h= 1e-4
fig = plt.figure(figsize=(12, 5))

E_AB = np.loadtxt('Adams-BashforthLE.txt', unpack=True)
E_AM = np.loadtxt('Adams-MoultonLE.txt', unpack=True)
E_V = np.loadtxt('VerletLE.txt', unpack=True)
E_RK4 = np.loadtxt('RK4NLE.txt', unpack=True)
E_FR = np.loadtxt('FRLE.txt', unpack=True)
t_AB = np.linspace(0, len(E_AB)*h, len(E_AB))
t_AM = np.linspace(0, len(E_AM)*h, len(E_AM))
t_V = np.linspace(0, len(E_V)*h, len(E_V))
t_RK4 = np.linspace(0, len(E_RK4)*h, len(E_RK4))
t_FR = np.linspace(0, len(E_FR)*h, len(E_FR))

def rel_E(E):
    return np.abs((E-E[0])/E[0])

plt.plot(t_RK4, rel_E(E_RK4), linewidth = 0.3, label='Runge-Kutta 4')
plt.plot(t_FR, rel_E(E_FR), linewidth = 0.3, label='Forest-Ruth')
plt.plot(t_V, rel_E(E_V), linewidth = 0.3, label='Verlet')
plt.plot(t_AB, rel_E(E_AB), linewidth = 0.3, label='Adams-Bashforth', alpha=0.7)
plt.plot(t_AM, rel_E(E_AM), linewidth = 0.3, label='Adams-Moulton', alpha=0.8)

plt.title('Long-term behaviour integrators (h=1e-4)')
plt.ylabel('|$\Delta$E/E|')
plt.xlabel('t (2.9722e9 s)')
plt.legend(loc='upper right')
plt.yscale('log')
```
